## Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
```

```python
plt.scatter(x,y1)
plt.show()
```

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
plt.scatter(x,y1,color='red', label="Apple", marker='^')
plt.scatter(x,y2,color='green', label="Kiwi")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plots - Prices Of Fruits Over Time")
plt.legend()
plt.show()
```

#### Adjust Size Of Any Plot

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y1,color='red', label="Apple", marker='^')
plt.scatter(x,y2,color='green', label="Kiwi")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plots - Prices Of Fruits Over Time")
plt.legend()
plt.show()
```

```python
themes = plt.style.available
plt.figure(figsize=(15,30))
for i, theme in enumerate(themes):
    plt.style.use(theme)
    plt.subplot(10, 3, i+1)
    plt.scatter(x,y1,color='r')
    plt.scatter(x,y2,color='b')
    plt.title(theme)
plt.show()
```
