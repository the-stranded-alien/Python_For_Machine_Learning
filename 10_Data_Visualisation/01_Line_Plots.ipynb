## Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.plot(x,y1,color='red')
plt.show()
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.xlabel("Time")
plt.ylabel("Price")
plt.show()
```

```python
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices Of Fruits Over Time")
plt.show()
```

```python
plt.plot(x,y1,color='red', label="Apple")
plt.plot(x,y2,color='green', label="Kiwi")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices Of Fruits Over Time")
plt.legend()
plt.show()
```

```python
plt.plot(x,y1,color='red', marker='o')
plt.plot(x,y2,color='green',linestyle="dashed",marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices Of Fruits Over Time")
plt.show()
```

#### If You Don't Have Y

```python
prices = np.array([1,2,3,4])**3
plt.plot(prices) # Indices Of Array Become Y
plt.show()
```

### Themes

```python
themes = plt.style.available
print(themes)
```

```python
plt.figure(figsize=(15,30))
for i, theme in enumerate(themes):
    plt.style.use(theme)
    plt.subplot(10, 3, i+1)
    plt.plot(x,y1,color='r')
    plt.plot(x,y2,color='b')
    plt.title(theme)
plt.show()
```
