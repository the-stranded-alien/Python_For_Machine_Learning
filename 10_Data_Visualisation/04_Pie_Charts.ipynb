## Pie Charts

```python
import matplotlib.pyplot as plt
```

```python
subjects = ["Maths", "Chemistry", "Physics", "English"]
weightage = [20, 10, 15, 5]

plt.pie(weightage, labels=subjects, explode=(0.05,0,0,0))
plt.show()
```

```python
plt.pie(weightage, labels=subjects, explode=(0.05,0,0,0), autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(weightage, labels=subjects, explode=(0.05,0,0,0), autopct='%1.1f%%',colors=["red","blue","green","yellow"])
plt.show()
```

```python
plt.pie(weightage, labels=subjects, explode=(0.05,0,0,0), autopct='%1.1f%%',shadow=True,colors=["red","blue","green","yellow"])
plt.show()
```

```python
plt.style.use("seaborn")
plt.pie(weightage, labels=subjects, explode=(0.05,0,0,0), autopct='%1.1f%%',shadow=True,colors=["red","blue","green","yellow"])
plt.show()
```

```python
plt.style.use("dark_background")
plt.pie(weightage, labels=subjects, explode=(0.05,0,0,0), autopct='%1.1f%%',shadow=True,colors=["red","blue","green","yellow"])
plt.show()
```
