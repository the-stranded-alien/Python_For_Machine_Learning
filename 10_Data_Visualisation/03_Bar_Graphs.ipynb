## Bar Graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,15], width=0.5) # Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12], width=0.5) # Next Year Prices
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width=0.5, label="Current Year", tick_label=["Gold", "Platinum", "Silver"]) # Current Year Prices
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5, label="Next Year") # Next Year Prices
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width=0.5, label="Current Year",color='blue') # Current Year Prices
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5, label="Next Year",color='orange') # Next Year Prices
plt.xticks(ticks=x_coordinates, labels=["Gold", "Platinum", "Silver"])
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width=0.5, label="Current Year") # Current Year Prices
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5, label="Next Year") # Next Year Prices
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.xticks(ticks=x_coordinates, labels=["Gold", "Platinum", "Silver"])
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width=0.5, label="Current Year") # Current Year Prices
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5, label="Next Year") # Next Year Prices
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-1,5)
plt.xticks(ticks=x_coordinates, labels=["Gold", "Platinum", "Silver"])
plt.legend()
plt.show()
```
