### Normal Distribution

- Prob. Distribution => Mean & Variance
- Standard Normal Distribution => Mean = 0 & Variance = 1
    - **X_std = ( X_ - Mean ) / Var**
    - **x_ = ( X_std * Var ) + Mean**

```python
import numpy as np
```

```python
Xsn = np.random.randn(100)
print(Xsn)
```

```python
sigma = 5
u = 60
X = np.round((Xsn*sigma) + u)
print(X)
```

## Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn")
plt.hist(X)
plt.xlabel("Marks Range")
plt.ylabel("Prob./Freq. Count Of Students")
plt.show()
```

```python
sigma = 8
u = 70
X = np.round((Xsn*sigma) + u)
X2 = np.round((Xsn*5) + 40)
```

```python
plt.hist(X,alpha=0.8)
plt.hist(X2,alpha=0.8)
plt.xlabel("Marks Range")
plt.ylabel("Prob./Freq. Count Of Students")
plt.show()
```

```python
plt.hist(X,alpha=0.8,label="Physics")
plt.hist(X2,alpha=0.8,label="Maths")
plt.xlabel("Marks Range")
plt.ylabel("Prob./Freq. Count Of Students")
plt.legend()
plt.show()
```
